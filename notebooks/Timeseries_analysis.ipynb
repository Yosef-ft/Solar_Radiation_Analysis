##### The Impact of cleaning on different columns

##### Impact of cleaning for Benin-Malanville dataset

```python
impact_A, impact_B, impact_WS, impact_WSgust = cleaning_impact(benin)
```

```python
print('Benin-malanville')
print(f"Impact on ModA: {impact_A} W/m")
print(f"Impact on ModB: {impact_B} W/m")
print(f"Impact on WS: {impact_WS} m/s")
print(f"Impact on WSgust: {impact_WSgust} m/s")
```

**Obeservation**
- Mean Differences: impact_A and impact_B are significantly positive, it indicates that cleaning likely improved the sensor readings.
- Cleaning sensors affects ModA and ModB more than cleaning sensors WS and WSgust.
- Time Series Visualization: We observe a clear jump or change in the trends after cleaning events, this visually confirms the impact.
